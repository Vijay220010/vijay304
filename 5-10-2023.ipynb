```python
import numpy as np
import pandas as pd 

from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

```python
df=pd.read_csv("C:/Users/Shubhamparamanik/OneDrive/Desktop/Birthweight_reduced_kg_R.csv")
df
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df,x='smoker',y='Birthweight')
plt.show()
```

```python
sample_01 = df[(df['smoker']==0)]
sample_02 = df[(df['smoker']==1)]
```

```python
sample_01.shape, sample_02.shape
```

```python
sample_02 = sample_02.sample(20)
sample_01.shape, sample_02.shape
```

```python
stats.levene(sample_01['Birthweight'], sample_02['Birthweight'])
```

```python
stats.ttest_ind(sample_01['Birthweight'], sample_02['Birthweight'])
```
