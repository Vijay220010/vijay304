```python
#importing numpy
import matplotlib.pyplot as plt
import numpy as np 
#create vector using numpy
v1 = np.array([[12],[15]])
v2 = np.array([[30],[40]])
v3 = np.array([[60],[70]])
#display the array
display(v1,v2,v3)
```

```python
#plotting the 3 vectors with matplotlib.quivery
origin=[0],[0]
v1=[12],[15]
v2=[30],[40]
v3=[60],[70]
plt.quiver(*origin, *v1,color='r', units='xy', angles='xy', scale_units='xy', label='v1=[10,15]')
plt.quiver(*origin, *v2,color='b', units='xy', angles='xy', scale_units='xy', label='v2=[98,40]')
plt.quiver(*origin, *v3,color='r', units='xy', angles='xy', scale_units='xy', label='v3=[34,70]')
#units , angles, scale_units and scale are used to scale the graph,below ,
plt.xlim(-8, 8)
plt.ylim(-9, 8)
plt.legend(loc=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title("3d represention of vectors")
```

```python
#create vector using numpy
origin=[0],[0]
v1=[5],[8]
v2=[-22],[32]
v3=[23],[43]
#display the array
print(v1)
print(v2)
print(v3)
#ploting vector
plt.quiver(*origin,*v1,color='r',units='xy',angles='xy',scale_units='xy',scale=1,label='v1=[15],[80]')
plt.quiver(*origin,*v2,color='b',units='xy',angles='xy',scale_units='xy',scale=1,label='v2=[-40],[79]')
plt.quiver(*origin,*v3,color='g',units='xy',angles='xy',scale_units='xy',scale=1,label='v3=[32],[43]')
#units,angles,scale_units and scale are used to scale the graph
plt.xlim(-8,8)
plt.ylim(-8,8)
#setting location for legend
plt.legend(loc=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title("3d representation of vectors")
#saving image as a png file
plt.savefig('3dvectors.png')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(8,8))
origin=[0],[0],[0]
v1=[8,[6,[0]
v2=[6,[4,[5
v3=[1],[3],[2]
ax=fig.add_subplot(111,projection='3d')
plt.quiver(*origin,*v1,color='k',label='v1[2,1,0]',arrow_length_ratio=0.2)
plt.quiver(*origin,*v2,color='g',label='v2[2,4,3]',arrow_length_ratio=0.5)
plt.quiver(*origin,*v3,color='r',label='v3[1,3,2]',arrow_length_ratio=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.set_zlim(-6,6)
plt.legend(loc=3)
plt.title("2d representation of vectors",color='b',size=25,family='magneto')
plt.show()
```
